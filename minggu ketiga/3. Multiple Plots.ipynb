```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, (ax1,ax2,ax3) = plt.subplots(1,3, sharey=True,figsize = (18,9))
ax1.plot(x, tan, 'b', label ='tan')

ax2.plot(x, sin, 'r', label ='sin')
ax2.set_title('Grafik sinx')

ax3.plot(x, cos, 'g', label ='cos')
ax3.set_title('Grafik cosx')

fig.legend(loc='upper center' )

# Hide x labels and tick labels for top plots and y ticks for right plots.
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize=(20,15))
ax.plot(x,y)

ins1 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins2 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins3 = ax.inset_axes([0.9,-0.1, 0.3, 0.3])
ins4 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])

ins1.plot(x,y5,'y',x,y,'blue')
ins2.plot(x,y2, 'r', x, y, 'blue')
ins3.plot(x,y3, 'g', x,y, 'blue')
ins4.plot(x, y4, 'black', x, y, 'blue')

ax.set_title('Grafik CosX')
ins1.set_title('Grafik CosX^5')
ins2.set_title('Grafik CosX^2')
ins3.set_title('Grafik CosX^3')
ins4.set_title('Grafik CosX^4')
```
